```python
@latexify.with_latex
def FormulaEcuacionCuadratica(a, b, c):
    return (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
FormulaEcuacionCuadratica    
```

**Supongamos que n especies componen una comunidad Σ en la que no hay interacciones interespecíficas. Este ecosistema puede ser modelado por:**

```python
@latexify.with_latex
def DinamicaCrecimiento(N, K, Lambda, i):
  return (Lambda * (i)) * N**i*((1-N**i/K(i))) 
DinamicaCrecimiento
```

```python
@latexify.with_latex
def LotkaVolterra(x, y, alpha, beta, delta, gamma):
  #X = x * (alpha - beta * y)
  return  x * (alpha - beta * y), y * (-delta + gamma * x)
LotkaVolterra   
```

```python
alpha, beta, delta, gamma, x0, y0 = 1., 1., 1., 1., 4., 2.
def LotkaVolterraArray(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = x * (alpha - beta * y)
    doty = y * (-delta + gamma * x)
    return np.array([dotx, doty])
Nt = 1000
tmax = 30.
t = np.linspace(0.,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(LotkaVolterraArray, X0, t, args = (alpha, beta, delta, gamma))
x, y = res.T

with plt.xkcd():
  plt.figure()
  plt.grid()
  plt.title("odeint method"); 
  plt.plot(t, x, 'xb', label = 'Palomas')
  plt.plot(t, y, '+r', label = "Rapaces")
  plt.xlabel('Time t, [days]')
  plt.ylabel('Population')
  plt.legend()
plt.show()
```
